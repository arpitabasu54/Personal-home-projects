```python
df1= sns.load_dataset('tips')
df1
```

```python
#2-------Relplot

sns.relplot(data= df1, x='tip',y='total_bill',hue='sex',col='time')
```

```python
sns.set_style('white')
sns.relplot(data= df1, x='total_bill',y='tip',hue='sex',col='day',col_order=['Fri','Sat','Sun','Thur'],col_wrap=2)
```

```python
#3----Barplot

sns.barplot(data= df1, x='total_bill',y='day',ci=None,hue= 'sex', palette= 'summer') 

#Ci is set at None to remove the black line which appears on the top of the bars
```

```python
## To create multiple plots together, Sir taught Facetgrid
s= sns.FacetGrid(df1, col= 'sex', height=4, palette= '') #height to make the fig size bigger
s.map(sns.barplot,'time', 'total_bill',ci=None)
```

```python
#4----Heatmap ( very much used in Data analysis)
import seaborn as sns
df= sns.load_dataset('car_crashes')

df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(),annot= True,cmap='RdBu') #cmap stands for colormapping of Red and Blue
plt.title('heatmap car_crashes')
```

```python
#5.----- distplot  ## for density

sns.set_style('darkgrid')
sns.distplot(df.speeding)
df.speeding.kurt()  #either skewness or kurtosis values are shown
df.speeding.skew()
```

```python
#6.----- histplot
sns.histplot(df.speeding)
sns.set_style('whitegrid')
```
